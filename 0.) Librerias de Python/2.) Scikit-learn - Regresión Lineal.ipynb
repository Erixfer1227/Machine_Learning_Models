**2.) Importar Librerias**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from regressors import stats
```

**3.) Cargar DataFrame**

```python
df = pd.read_excel('/content/2.) homework & ach.xls')
```

**5.) Preprocesamiento de Datos**

**5.1) Asignar independientes y dependientes**

```python
x = df[['Horas de Estudio']]
```

```python
x.head(1)
```

```python
y = df[['Calificación']]
```

```python
y.head(1)
```

**6.) Análisis exploratorio de Datos (Gráficas)**

```python
# Plotear gráfco de dispersión
plt.scatter(x,y)

# Definir ejes
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificación")
plt.title("ScatterPlot Hrs. Estudio Vs Calificaciones")

# Poner cuadricula en el gráfico
plt.grid()
```

**7.) Creando el Modelo de Regresión Lineal**

**7.1) Definir el Modelo a Utilizar**

```python
modelo = LinearRegression()
```

**7.2) Entrenar el modelo (Función Fit)**

```python
# Entrenar el Modelo con los datos de X y los datos de Y
modelo.fit(x, y)
```

**7.3) Evaluar la Calidad del Modelo**

```python
# Resultados de la Regresión
print("==========Resumen Estadistico==========")
stats.summary(modelo, x, y)

# Intercepto y variable independiente son estadisticamente significativos |T| > 2 y p < 0.05
# Las horas solo explican en 10% la calificacion de un alumno
```

```python
# Código para crear el gráfico
plt.plot(x, modelo.predict(x), label="predicciones")

# Agregar un título y etiquetas de ejes
plt.title("Regresión Lineal Simple")
plt.xlabel("Horas")
plt.ylabel("Calificación")

# Agregar una leyenda y mostrar la cuadrícula
plt.legend(loc="best")  # "best" coloca la leyenda en la mejor ubicación
plt.grid()

# Mostrar el gráfico
plt.show()
```

**7.4) Predecir (con función predict)**

```python
xi = 3
yi = modelo.predict([[xi]]) # <--- Ingresar input del modelo (Ej. Estudio 3 horas)
resultado = ("El estudiante obtendra una calificación de:", yi, "si estudia:", xi, "horas" )
print(resultado)
```
