## **Transformas Datos o Ingenier√≠a de Carateristicas**

```python
# Crear una nueva columna

df_llenado = df.fillna({"retweets":80, "followees": 1190.185185 , "followers": 352.807692 }) 


# 1.) Crear funci√≥n que generar√° los valores de la nueva columan
def calcularGanancias(retweets): # La funci√≥n se llama calcularganancias y el input de la funci√≥n son los valores de retweets
    ganancia = retweets * 4 #La variable ganancia almacena el producto de cada valor de retweets por cuatro
    return ganancia #La salida de la funci√≥n esta en la variable ganancia

# 2.) Crear nueva columna
df_llenado["ganancia"] = df_llenado["retweets"].apply(calcularGanancias)
# df["Nombre de la nueva columa"] = df ["columna a la que se aplica la funci√≥n"].apply(funci√≥n a aplicar)
# por cada valor de la columna input aplica la funci√≥n tal ...

df_llenado.head(3)
```

```python
# Crear una nueva columna con dos variables del dataframe

# 1.) Crear funci√≥n que generar√° los valores de la nueva columan
def popularidad(fila): 
    resultado = fila["followees"]/fila["followers"]
    return resultado 

# 2.) Crear nueva columna
df_llenado["popularidad"] = df_llenado.apply(popularidad, axis=1)

df_llenado.head() #Menor a 1 popular
```

## **Agrupaci√≥n y Agregaci√≥n de Datos**

```python
df_llenado.groupby("country").agg({
    "followers": 'sum',
    "mentions": 'mean',
    "retweets": 'max'
})
```

```python
grouped = df_llenado.groupby("country").agg({
    "followers": 'sum',
    "mentions": 'mean',
    "retweets": 'max'
})

grouped[grouped["followers"]>5000]
```
